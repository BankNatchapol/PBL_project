# Transform sensing data into csv 

```python
f = []
for (dirpath, dirnames, filenames) in os.walk('/content/gdrive/MyDrive/data/71_sensing_data'):
  n = []
  for i in filenames:
    if i[-3:] != 'csv':
      continue
    n.append(os.path.join(dirpath , i))
  f.extend(n)

def find_nth(haystack, needle, n):
    start = haystack.find(needle)
    while start >= 0 and n > 1:
        start = haystack.find(needle, start+len(needle))
        n -= 1
    return start
s = set()
for i in f:
  c = i[63:]
  a = find_nth(c, '_', 1)
  b = find_nth(c, '_', 2)
  s.add(c[a+1:b])

for d in s:
  exec(d + " = pd.DataFrame()")

for file in f:
  c = file[63:]
  a = find_nth(c, '_', 1)
  b = find_nth(c[a+1:], '_', 1)
  
  if c[a+1:a+b+1] == 'activ':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    activ = activ.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'exist':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    exist = exist.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'sphy':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    sphy = sphy.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'temp':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    temp = temp.append(tem, ignore_index=True)
  if c[a+1:a+b+1] == 'weight':
    tem = pd.read_csv(file)
    tem['SubjectID'] = c[:a]
    weight = weight.append(tem, ignore_index=True)
    
activ['SubjectID'] = activ['SubjectID'] + '01'
exist['SubjectID'] = exist['SubjectID'] + '01'
sphy['SubjectID'] = sphy['SubjectID'] + '01'
temp['SubjectID'] = temp['SubjectID'] + '01'
weight['SubjectID'] = weight['SubjectID'] + '01'    
activ.to_csv('activ.csv', index = False)
exist.to_csv('exist.csv', index = False)
sphy.to_csv('sphy.csv', index = False)
temp.to_csv('temp.csv', index = False)
weight.to_csv('weight.csv', index = False)
```
